フォルダ確認

```python
import os

# ベースフォルダパス（番号部分を除いた固定部分）
base_folder_path = r"C:\Users\scrum\OneDrive\ドキュメント\lab\実験\被験者データ"

# フォルダの番号範囲を指定（01, 02, ..., 25）
for i in range(1, 26):  # 必要な番号の範囲を設定（例: 1〜25）
    # フォルダ番号を2桁にする（01, 02, ...）
    folder_number = f"{i:02d}"
    # 完全なフォルダパスを生成
    folder_path = os.path.join(base_folder_path, folder_number, "logs")
    
    print(f"\n確認中: '{folder_path}'")
    if os.path.exists(folder_path):
        print(f"'{folder_path}' 内のファイルとフォルダ:")
        # フォルダ内のファイルやディレクトリをリストアップ
        for item in os.listdir(folder_path):
            item_path = os.path.join(folder_path, item)
            if os.path.isdir(item_path):
                print(f"📁 フォルダ: {item}")
            elif os.path.isfile(item_path):
                print(f"📄 ファイル: {item}")
    else:
        print(f"指定したパス '{folder_path}' は存在しません。")
```
